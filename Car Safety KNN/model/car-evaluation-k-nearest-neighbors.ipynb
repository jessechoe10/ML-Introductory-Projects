```python
car_data = pd.read_csv('../input/car-evaluation-data-set/car_evaluation.csv')
car_data.columns = ['price', 'maint', 'doors', 'ppl', 'lug_boot', 'safety', 'acceptable']
car_data
```

```python
from sklearn import linear_model, preprocessing
from sklearn.model_selection import train_test_split

le = preprocessing.LabelEncoder()

price = le.fit_transform(list(car_data['price']))
maint = le.fit_transform(list(car_data['maint']))
doors = le.fit_transform(list(car_data['doors']))
ppl = le.fit_transform(list(car_data['ppl']))
lug_boot = le.fit_transform(list(car_data['lug_boot']))
safety = le.fit_transform(list(car_data['safety']))
acceptable = le.fit_transform(list(car_data['acceptable']))

predict = "acceptable"

X = list(zip(price, maint, doors, ppl, lug_boot, safety, acceptable))
y = list(acceptable)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 9)
model.fit(X_train, y_train)

percent_acc = model.score(X_test, y_test) * 100
print("Accuracy: %3.f" % percent_acc + "%")

predicted = model.predict(X_test)
names = ["acc", "good", "unacc", "vgood"]

for x in range(len(predicted)):
    print("Predicted:", names[predicted[x]], " ", "Actual:", names[y_test[x]])
```
