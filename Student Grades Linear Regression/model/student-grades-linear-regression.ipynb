```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

grades = pd.read_csv('../input/student-performance-data-set/student/student-mat.csv', sep = ';')
grades.head()
```

```python
plt.figure(figsize = (14, 6))
plt.title("Quarter 1 Grades vs Quarter 2 Grades")
sns.lineplot(x = grades['G1'], y = grades['G2'])
plt.xlabel('Quarter 1 Grades')
plt.ylabel('Quarter 2 Grades')
plt.savefig('q1vq2.png')
```

```python
plt.figure(figsize = (14, 6))
plt.title("Quarter 2 Grades vs Quarter 3 Grades")
sns.lineplot(x = grades['G2'], y = grades['G3'])
plt.xlabel('Quarter 2 Grades')
plt.ylabel('Quarter 3 Grades')
plt.savefig('q2vq3.png')
```

```python
plt.figure(figsize = (14, 6))
plt.title("Quarter 1 Grades vs Quarter 3 Grades")
sns.lineplot(x = grades['G1'], y = grades['G3'])
plt.xlabel('Quarter 1 Grades')
plt.ylabel('Quarter 3 Grades')
plt.savefig('q1vq3.png')
```

```python
def plot(slope, intercept):
    x_points = [] 
    y_points = []
    i = 0.0
    while(i <= 20.0):
        x_points.append(i)
        y_points.append(slope * i + intercept)
        i += 0.1
    return x_points, y_points
```

```python
x = grades['G2']
y = grades['G3']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 0)
x_test = x_test.values.reshape(-1, 1)
x_train = x_train.values.reshape(-1, 1)
lin_reg = linear_model.LinearRegression()
lin_reg.fit(x_train, y_train)
m = np.array(lin_reg.coef_)[0]
b = lin_reg.intercept_
pred = lin_reg.predict(x_test)
print("y = %.2f" % m, 'x + %.2f' % b)
percent_acc = 100.0 * r2_score(y_test, pred)
print("Accuracy: %3.f" % percent_acc + "%")
plt.figure(figsize = (14, 6))
plt.title("Quarter 2 Grades vs Quarter 3 Grades")
sns.lineplot(x = grades['G2'], y = grades['G3'], label = 'Actual')
plt.xlabel('Quarter 2 Grades')
plt.ylabel('Quarter 3 Grades')
x_vals, y_vals = plot(m, b)
sns.lineplot(x = x_vals, y = y_vals, label = 'Predicted')
plt.savefig('q2vq3.png')
```
